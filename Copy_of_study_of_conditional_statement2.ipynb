```python
# Check whether number is positive or negative
number = input("Enter a number: ")
if int(number) >= 0:
    print("Number is positive")
else:
      print("Number is negative")
```

```python
# Check whether number is positive or negative using elif
number = int(input("Enter a number: "))
if number > 0:
    print("Number is positive")
elif number < 0:
      print("Number is negative")
else:
  print("Number is 0")
```

```python
# Even odd Numbers
number = int(input("Enter a number: "))
if number % 2 == 0:
  print("Number is even")
else:
  print("Number is odd")
```

```python
# Find largest among 2 number
number1 = int(input("Enter first number:"))
number2 = int(input("Enter second number:"))
if number1 > number2:
  print("First Number is greater than Second Number ")
else:
  print("Second Number is greater than First Number")
```

```python
# Largest number among 3 numbers
number1 = int(input("Enter first number:"))
number2 = int(input("Enter second number:"))
number3 = int(input("Enter third number:"))
if number1 > number2 and number1 > number3:
  print("First Number is greater than Second Number and Third Number")
elif number2 > number1 and number2 > number3:
  print("Second Number is greater than First Number and Third Number")
else:
  print("Third Number is greater than First Number and Second Number")
```

```python
# Take marks of 5 subjects and display grade using average
print("Enter marks out of 100.")
sub1 = int(input("Enter marks of first subject:"))
sub2 = int(input("Enter marks of second subject:"))
sub3 = int(input("Enter marks of third subject:"))
sub4 = int(input("Enter marks of fourth subject:"))
sub5 = int(input("Enter marks of fifth subject:"))

total_marks = sub1 + sub2 + sub3 + sub4 + sub5
average = total_marks / 5

print("Average Percentage:", average)
if average >= 90:
  print("Grade: O")
elif average >= 80:
  print("Grade: A+")
elif average >= 70:
  print("Grade: A")
elif average >= 60:
  print("Grade: B+")
elif average >= 50:
  print("Grade: B")
elif average >= 40:
  print("Grade: pass")
else:
  print("Grade: F")
```
